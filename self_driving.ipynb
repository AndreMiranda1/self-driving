## Cleverti Talks - Self Driving Race Car

In this talk we'll explore how self driving cars work, by training a Convolution Neural Network to steer a wheel at a given time during a track race.

We used a Kaggle Dataset with pictures from a virtual race track, where each picture is labeled with a steering angle.
https://www.kaggle.com/roydatascience/training-car

In the end we'll test the model with a real video (not present in the training data!) from a race available in Youtube.

```python
import pandas as pd
from keras import models, layers, losses
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Processing images and building training data
img = cv2.imread('data/IMG/center_2016_12_01_13_30_48_287.jpg',0)
img_size = 150
img = cv2.resize(img, (img_size, img_size))
plt.imshow(img, cmap='gray')

img_data = []
target_data = []
for image in list(df['center']):
    img = cv2.imread(os.path.join('data', image), 0)
    img = cv2.resize(img, (img_size, img_size))
    img = img.reshape((img_size, img_size, 1))
    
    steering_value = df[df['center'] == image].iloc[0]['steering']
    
    if abs(steering_value) < 0.1:
        img_data.append(img)
        target_data.append(steering_value)
    else:
        # balancing dataset
        for i in range(7):
            img_data.append(img)
            target_data.append(steering_value)

img_data = np.asarray(img_data)
img_data= img_data.astype('float32') / 255
display(img_data.shape)
```

```python
# Building the model
def build_model():
    model = models.Sequential()
    # Convolution NN part
    model.add(layers.Conv2D(32, (5,5), activation='relu', input_shape=(img_size, img_size, 1)))
    model.add(layers.MaxPooling2D((3, 3)))
    model.add(layers.Conv2D(64, (5,5), activation='relu'))
    model.add(layers.MaxPooling2D((3, 3)))
    model.add(layers.Conv2D(64, (5,5), activation='relu'))

    # Classifier Part
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))

    # Tunning the model
    model.compile(optimizer='rmsprop',  # a standard optimizer
                  loss='mse',  # the suitable for regression problems
                  metrics=['mae'])
    return model

nr_train_samples = int(round(len(img_data)*0.9, 0))

train_data, validation_data = img_data[:nr_train_samples], img_data[nr_train_samples:]
train_targets, validation_targets = target_data[:nr_train_samples], target_data[nr_train_samples:]

load_model = True
if not load_model:
    model = build_model()
    nr_epochs = 3
    history = model.fit(train_data, train_targets, epochs=nr_epochs, batch_size=350,
                        validation_data=[validation_data, validation_targets])
    # Monitor the Loss
    history_dict = history.history
    loss_values = history_dict['loss']
    val_loss_values = history_dict['val_loss'] 
    # Plotting the Performance on each epoch
    epochs = range(1, nr_epochs + 1)
    plt.plot(epochs, loss_values, 'bo', label='Training Loss')
    plt.plot(epochs, val_loss_values, 'b', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
else:
    model = models.load_model('driving_model.h5')
```

```python
# Test on a dataset image
img = cv2.imread('data/IMG/center_2016_12_01_13_31_12_937.jpg',0)
img = cv2.resize(img, (img_size, img_size))
plt.imshow(img, cmap='gray')

img = img.reshape((1, img_size, img_size, 1))
img= img.astype('float32') / 255
display(model.predict(img))
```

```python
def rotateImage(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result

wheel = cv2.imread('wheel.png',0)
plt.imshow(wheel, cmap='gray')

rotate_wheel = rotateImage(wheel, 25)
plt.imshow(rotate_wheel, cmap='gray')
```

```python
cap = cv2.VideoCapture('video_test.mp4')
from time import sleep
from random import random

while True:
    if cap.grab():
        flag, frame = cap.retrieve()
        if not flag:
            continue
        else:
            cv2.imshow('video', frame)
            
            # Processing and predicting
            img_to_predict = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            img_to_predict = cv2.resize(img_to_predict, (img_size, img_size))

            plt.imshow(img_to_predict, cmap='gray')

            img_to_predict= img_to_predict.astype('float32') / 255
            plt.imshow(img_to_predict, cmap='gray')
            img_to_predict = img_to_predict.reshape((1, img_size, img_size, 1))
            steering = model.predict(img_to_predict)
            
            rotate_wheel = rotateImage(wheel, -steering[0][0]*180)
            cv2.imshow('wheel', rotate_wheel)
            
    if cv2.waitKey(10) == 27:
        break
    sleep(0.05)
 
cap.release()
cv2.destroyAllWindows()
```
